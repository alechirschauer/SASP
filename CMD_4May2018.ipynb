```python
# import stuff you will use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# the command below is a jupyter notebook magic command
# matplotlib will automatically show plots inline
# if you dont do this you will have to do plt.show()
%matplotlib inline
```

```python
phot_data = pd.read_csv('CMDparameters27April2018_ahirschauer.csv')
```

```python
phot_data['threesixMINUSfiveeight'] = phot_data.k36mag.values - phot_data.k58mag.values
phot_data['fourfiveMINUSfiveeight'] = phot_data.k45mag.values - phot_data.k58mag.values
phot_data['fourfiveMINUSeightzero'] = phot_data.k45mag.values - phot_data.k80mag.values
phot_data['fourfiveMINUStwentyfour'] = phot_data.k45mag.values - phot_data.k24mag.values
phot_data['eightzeroMINUStwentyfour'] = phot_data.k80mag.values - phot_data.k24mag.values
```

```python
filter_phot_data = phot_data[(phot_data.threesixMINUSfiveeight > -2.5) & (phot_data.threesixMINUSfiveeight < 4.5)]
```

```python
# make a figure and set the size, you dont have to set tge size
plt.figure(figsize=(8,8))
# Still a good idea to set the limit
plt.xlim(-2.0,5.0)
plt.ylim(21.0,8.0)
# Before you weren't plotting the filtered data
plt.plot(filter_phot_data.threesixMINUSfiveeight,filter_phot_data.k36mag,'.')
plt.ylabel('[3.6]')
plt.xlabel('[3.6]-[5.8]')
plt.title('[3.6]-[5.8] vs. [3.6]')
```

```python
# Wanted to make sure I could plot two side by side
plt.figure(figsize=(12,6))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

ax1.set_ylim(21.0,8.0)
ax1.set_xlim(-2.0,5.0)
ax1.plot(filter_phot_data.threesixMINUSfiveeight,filter_phot_data.k36mag,'.')
ax1.set_ylabel('[3.6]')
ax1.set_xlabel('[3.6]-[5.8]')
ax1.set_title('[3.6]-[5.8] vs. [3.6]')

ax2.set_ylim(21.0,8.0)
ax2.set_xlim(-2.0,5.0)
ax2.plot(filter_phot_data.threesixMINUSfiveeight,filter_phot_data.k36mag,'.')
ax2.set_ylabel('[3.6]')
ax2.set_xlabel('[3.6]-[5.8]')
ax2.set_title('[3.6]-[5.8] vs. [3.6]')
```
